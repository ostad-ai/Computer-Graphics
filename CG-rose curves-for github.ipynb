## Rose curves
In *polar coordinates*, every point is represented by $(r,\theta)$
<br>A **rose curve** is defined in *polar coordinates* by
- $r=a\,cos(k\theta)$

Generally, $k$ and $a$  are constant real numbers.
<br>To represent a rose curve in *Cartesian coordinates*: 
 - $x=r\,cos(\theta)=a\,cos(k\theta)\,cos(\theta)$
 - $y=r\,sin(\theta)=a\,cos(k\theta)\,sin(\theta)$
 
**Example:** For $k=1$, we get to the curve of a *circle*:
- $r(\theta)=a\,cos(\theta)$, where $a$ is the diameter of the circle.

```python
# importing required modules
from matplotlib import pyplot as plt
import numpy as np
from math import pi,gcd
```

```python
# returns the (x,y) points of rose curve:
# r=a*cos((n/d)*teta)
# we assume n and d are coprime
# which means that GCD(n,d)=1
def rosend(n=7,d=5,a=10,Npoints=500):
    if n%2==0 or d%2==0:
        cycle=2*d*pi
    else:
        cycle=d*pi
    tetas=np.linspace(0,cycle,Npoints)
    r=a*np.cos(n/d*tetas)
    xs=r*np.cos(tetas)
    ys=r*np.sin(tetas)
    return xs,ys
#-------------------------------
# displaying a number of rose curves with different n and d
plt.figure(figsize=(16,22))
ns=list(range(1,8))
ds=list(range(1,8))
i=0
for n in ns:
    for d in ds:
        if gcd(n,d)==1:
            i+=1
            xs,ys=rosend(n,d)
            plt.subplot(6,6,i)
            plt.plot(xs,ys,'-')
            plt.title(f'n={n}, d={d}')
plt.suptitle(r'Rose curve: $r=a\,cos(\frac{n}{d} \theta)$',y=.92,size=20)
plt.show()
```
