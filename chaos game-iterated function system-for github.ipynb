```python
import random
from matplotlib import pyplot as plt
```

```python
class Point:
    def __init__(self,x,y):
        self.x=x; self.y=y
        
def chaosGame(Ts,Probs,iter=10000,
    palette=['red','blue','green'],discard=20):
    P=Point(0,0)
    points=[]
    for i in range(iter):
        index=random.choices(range(len(Ts)),Probs)[0]
        T=Ts[index]
        P.x,P.y=T[0]*P.x+T[1]*P.y+T[4],T[2]*P.x+T[3]*P.y+T[5]
        points.append((P.x,P.y,palette[index]))
    return points[discard:]
```

```python
#-----spiral 
Ts=[[.787879,-.424242,.242424,.859848,1.758647,1.408065],
    [-.121212,.257576,.151515,.053030,-6.721654,1.377236],
   [.181818,-.136364,.090909,.181818,6.086107,1.568035]]
Probs=[.895652,.052174,.052174]
points=chaosGame(Ts,Probs,50000)
xs,ys,colors=zip(*points)
plt.scatter(xs,ys,c=colors,s=.1,alpha=.5)
plt.show()
```
