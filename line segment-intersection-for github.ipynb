## Intersections between Line Segments
A line segment can be defined in *parametric form* by its two endpoints $P_1$ and $P_2$:
<br> $P(t)=P_1+t(P_2-P_1)$, $t\in[0, 1]$
<br> Where $t$ is the *parameter* of the line segment.
<br>We can find the intersection of two line segments using parametric representation of them.
<br>The formulas for intersection point of line segments are given in the **Pinterest** page mentioned below.

```python
# importing required modules
import random
from matplotlib import pyplot as plt
```

```python
# a line segment between two endpoints p0 and p1
# p0 and p1 are expressed by tuple (x,y)
class LineSegment:
    def __init__(self,p0,p1,eps=.001):
        self.p0=p0 
        self.p1=p1
        self.eps=eps
        # finds intersection of this LS to other
    def intersect(self,other):
        x1,y1=self.p0
        x2,y2=self.p1
        x3,y3=other.p0
        x4,y4=other.p1
        denom=(x2-x1)*(y3-y4)+(x4-x3)*(y2-y1)
        if abs(denom)<self.eps: # lines are parallel
            return None
        s=((x3-x1)*(y3-y4)+(x4-x3)*(y3-y1))/denom
        t=((x2-x1)*(y3-y1)-(x3-x1)*(y2-y1))/denom
        if s>=0 and s<=1 and t>=0 and t<=1:
            inter_x=x1+s*(x2-x1)
            inter_y=y1+s*(y2-y1)
            return (inter_x,inter_y)
        else:
            return None #intersection is outside
```

```python
# example
WIDTH,HEIGHT=700,700
segs=[]  #holds the line segments
inters=[] # holds the intersection points

# generating some random line segments
for _ in range(12):
    x1,y1=random.randrange(0,WIDTH),random.randrange(0,HEIGHT)
    x2,y2=random.randrange(0,WIDTH),random.randrange(0,HEIGHT)
    segs.append(LineSegment((x1,y1),(x2,y2)))
    
# finding intersection points between line segments
for i in range(len(segs)):
        for j in range(i+1,len(segs)):
            inter=segs[i].intersect(segs[j])
            if inter!=None:
                inters.append(inter)
                
# display the line segment
for seg in segs:
    plt.plot([seg.p0[0],seg.p1[0]],[seg.p0[1],seg.p1[1]],'-')
    
# display the intersection points
for inter in inters:
    plt.text(inter[0],inter[1],'\U0001F608',va='center',ha='center')
plt.title('Line segments and their intersections')
plt.show()
```
